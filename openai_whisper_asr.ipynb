```python
import whisper
import jiwer
import pandas as pd
import os
import shutil
from tqdm import tqdm
```

```python
df = pd.read_csv('/content/extract_content/hindi/test/bucket.csv')
```

```python
df
```

```python
audio_folders = '/content/extract_content/hindi/test/audio'
audio_files = 'AudioFiles'

if not os.path.exists(audio_files):
    os.makedirs(audio_files)

for folder in os.listdir(audio_folders):
    folder_path = os.path.join(audio_folders, folder)

    for file in os.listdir(folder_path):
        source_path = os.path.join(folder_path, file)
        dest_path = os.path.join(audio_files, file)

        shutil.copy2(source_path, dest_path)
```

```python
df
```

```python
df.rename(columns={'transcript':'original_transcript'}, inplace=True)
```

```python
df = df.assign(whisper_transcript=None, WER=None)
```

```python
df['file_path'] = df['file_path'].astype(str)
df['file_path'] = df['file_path'].apply(lambda x: x.split('/')[-1])
```

```python
df
```

```python
df.rename(columns={'file_path':'file_code'}, inplace=True)
```

```python
model = whisper.load_model('medium')
```

```python
df
```

```python
file_name = '844424930324566-536-m.wav'
```

```python
df.loc[df['file_code'] == file_name, 'original_transcript'].values[0]
```

```python
import warnings
warnings.filterwarnings('ignore')


num_files = len(os.listdir(audio_files))
progress_bar = tqdm(total=num_files, desc="Processing Files", unit="file")

for file in os.listdir(audio_files):
    result = model.transcribe(os.path.join(audio_files, file), language='hi')
    df.loc[df['file_code'] == file, 'whisper_transcript'] = result['text']
    df.loc[df['file_code'] == file, 'WER'] = jiwer.wer(df.loc[df['file_code'] == file, 'original_transcript'].values[0], result['text'])

    progress_bar.update(1)

progress_bar.close()
```
